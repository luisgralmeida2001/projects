```python
import pandas as pd

df = pd.read_excel(r'G:\Drives compartilhados\USINAS\USINAS GESTAO\- Planejamento Estratégico Usinas\CNPJS.xlsx')

display(df.head(3))
```

```python
import requests
import json
import time

#Função para a consulta dos cnpjs
def consulta_cnpj(cnpj):
        url =  f"https://receitaws.com.br/v1/cnpj/{cnpj}"
        querystring = {"token": "XXXXXXXX-XXXX-XXXX-XXXX-XXXXXXXXXXX", "cnpj": cnpj, "plugin": "RF"}
        response = requests.get(url, params=querystring)

#Lista com os CNPJS criados
        try:
            resp = json.loads(response.text)
            cnpj_data = {
                "CNPJ": cnpj,
                "Nome": resp.get('nome', ''),
                "Email": resp.get('email', ''),
                "Número": resp.get('telefone', ''),
                "Situação": resp.get('situacao', ''),
                "Logradouro": resp.get('logradouro', ''),
                "Número": resp.get('numero', ''),
                "Município": resp.get('municipio', ''),
                "Bairro": resp.get('bairro', ''),
                "UF": resp.get('uf', ''),
                "CEP": resp.get('cep', ''),
                "Capital Social": resp.get('capital_social')
            }
            return cnpj_data
        except Exception as e:
            print(f"Erro ao consultar CNPJ {cnpj}: {str(e)}")
            return None
        
#Lista com os CNPJS
cnpjs = df['CNPJ'].tolist()

# Listas para armazenar os valores retornados
cnpj_list = []
nome_list = []
email_list = []
numero_list = []
situacao_list = []
logradouro_list = []
municipio_list = []
bairro_list = []
uf_list = []
cep_list = []
capital_social_list = []

# Iterar pelos CNPJs e preencher as listas
for cnpj in cnpjs:
    result = consulta_cnpj(cnpj)
    if result:
        cnpj_list.append(result["CNPJ"])
        nome_list.append(result["Nome"])
        email_list.append(result["Email"])
        numero_list.append(result["Número"])
        situacao_list.append(result["Situação"])
        logradouro_list.append(result["Logradouro"])
        municipio_list.append(result["Município"])
        bairro_list.append(result["Bairro"])
        uf_list.append(result["UF"])
        cep_list.append(result["CEP"])
        capital_social_list.append(result["Capital Social"])

        # Adicionar um atraso de 20 segundos entre as solicitações
        time.sleep(20)

        # Adicionar instruções de impressão para verificar os valores
        print(f"CNPJ: {result['CNPJ']}, Nome: {result['Nome']}")

# Imprimir as primeiras entradas para verificar
print(f"Primeiros 5 CNPJs: {cnpj_list[:5]}")
print(f"Primeiros 5 Nomes: {nome_list[:5]}")
```

```python
# Crie um DataFrame do Pandas com as listas
data = {
    "CNPJ": cnpj_list,
    "Nome": nome_list,
    "Email": email_list,
    "Número": numero_list,
    "Situação": situacao_list,
    "Logradouro": logradouro_list,
    "Município": municipio_list,
    "Bairro": bairro_list,
    "UF": uf_list,
    "CEP": cep_list,
    "Capital Social": capital_social_list
}
df = pd.DataFrame(data)

# Especifique o caminho completo do arquivo Excel
caminho_completo = r'G:\Drives compartilhados\USINAS\USINAS GESTAO\- Planejamento Estratégico Usinas\dados_empresas.xlsx'

# Salve o DataFrame como um arquivo Excel no caminho especificado
df.to_excel(caminho_completo, index=False)

print(f'Arquivo Excel "{caminho_completo}" criado com sucesso.')
```
